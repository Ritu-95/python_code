We have lots of null values . Lets Analyse further

```python
null_col = df.isnull().sum().sort_values(ascending = False)
null_col = null_col[null_col.values > (0.35*len(df))]
```

Plotting bar graph for null values greater than 35%

```python
plt.figure(figsize=(20,4))
null_col.plot(kind='bar', color="#4CB391")
plt.title('List of Columns & Null counts where null values are more than 35%')

plt.xlabel("Null Columns",fontdict={"fontsize":12,"fontweight":6})
plt.ylabel("Count of null values",fontdict={"fontsize":12,"fontweight":6})
plt.show()
```

```python
null_list_col =list(null_col.index.values)
df.drop(labels = null_list_col,axis=1,inplace= True)
```
